````python
from bs4 import BeautifulSoup
def parse_html_table(table):
    n_columns = 0
    n_rows = 0
    column_names = []
    columns =[]
    for idx,row in enumerate(table.find_all('tr')):
        td_tags = row.find_all('td')
        if len(td_tags) > 0:
            n_rows += 1
            if n_columns == 0:
                # Set the number of columns for our table
                n_columns = len(td_tags)
        th_tags = row.find_all('th')
        if len(th_tags) > 0 and len(column_names) == 0:
            for th in th_tags:
                column_names.append(th.get_text())
        row_values = []
        for td in td_tags:
            row_values.append(td.get_text().strip())
        columns.append(row_values)
    row_names = []
    if len(column_names)==0:
        for th in columns[0]:
            column_names.append(th)
        for d in columns[1:]:
            if not d:continue
            row_names.append(d[0])
    return [columns,row_names,column_names]
def convert_table(header, data):
    result = []
    for row in data[1:]:
        t_str = ''
        for i in range(1 if header[0].strip()=='' else 0,len(header)):
            if i<len(row):
                t_str+=header[i]+'的'+row[0]+'是'+row[i]+','
        t_str = t_str.strip(',')
        # print(t_str)
        result.append(t_str)
    result_str = ';'.join(result)
    return result_str
def read_html(filename):
  html_str = ''
  with open(filename,'r',encoding='utf-8') as lines:
    for line in lines:
      html_str+=line.strip()
  html = BeautifulSoup(html_str,parser='html.parser')
  all_text = ''
  idx = 1
  table_idx =0
  while idx<10000:
    id = 'hx_'+str(idx)
    div = html.find("div", {"id": id})
    table = html.find('table',{'id':id})
    if div:
      all_text += div.get_text()
    elif table:
      import copy
      t_table =copy.copy(table)
      d = parse_html_table(table)
      try:
        table_text = convert_table(d[2],d[0])
        
        all_text+=table_text.replace('{','').replace('}','').replace('[','').replace('}','').replace(']','')
          # all_text+=re.sub('class.*hx_(\d+)\"', '', repr(clean_html(t_table))).replace('<div>', '').replace('</div>', '')
      except Exception as e:
          # print(e)
        pass
      t_id = table.find_all('div')[-1].get('id').split('_')[1].strip()
      idx = int(t_id)
    idx+=1
  return all_text
````

````python
def get_new_prompt(context,query):
    prompt= """
    给定背景知识："""+context+"""
    您有以下工具可以使用：
    数学：调用此工具与 Wolfram Alpha API 进行交互。Wolfram Alpha API 在解答关于数学、科学、技术、文化、社会和日常生活的问题时非常有用。参数：[{"name": "query", "type": "string", "description": "要由 Wolfram Alpha 解决的问题", "required": true}] 将参数格式化为 JSON 对象。
    使用以下格式：
    问题：您必须回答的输入问题
    思考：您应该始终思考该如何处理，回答内容严格参考背景知识内容
    行动：需要采取的行动，应为 [Math] 之一
    行动输入：计算公式的输入，计算公式应保证正确，并将相应内容请替换为可计算的数值，格式如:{"query":(1+2)/3}
    观察：行动的结果
    ...（这个思考/行动/行动输入/观察可以重复零次或多次）
    思考：我现在知道了最终答案
    最终答案：原始输入问题的最终答案，请保证回答的准确性，如果根据背景无法回答，请返回无法回答
    问题:"""+query
    return prompt
````

````python
from sentence_transformers import SentenceTransformer
model_name = 'BAAI/bge-base-zh'
import faiss
m = SentenceTransformer(model_name,device='cuda')
faiss_index = None 
content = None 
input_filename = None 
import numpy as np 
import time 
def embed_sentences(sentences):
    sentence_embeddings = m.encode(sentences)
    return sentence_embeddings
def do_faiss_lookup(fastIndex, query_text, model, top_k):
  n_embeddings =768
  embedding_dim = 768 
  embedding_q = np.reshape(model.encode(query_text),[1,embedding_dim])
  embedding_q = embedding_q.astype('float32')
  faiss.normalize_L2(embedding_q)
  st = time.time()
  matched_em, matched_indexes = fastIndex.search(embedding_q, top_k) # it returns matched vectors and thier respective indexes, we are interested only in indexes.
  et = time.time()
  return matched_em, matched_indexes[0]
def read_pdf_and_search(filename):
    from langchain.document_loaders import PyPDFium2Loader
    from langchain.document_loaders import TextLoader
    file_path = filename
    if '.html' not in filename:
        loader = PyPDFium2Loader(file_path) if '.pdf' in file_path else TextLoader(file_path)
        documents = loader.load()
        from langchain.text_splitter import RecursiveCharacterTextSplitter
        text_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=100)
        split_documents = text_splitter.split_documents(documents)
    else:
        html_text = read_html(file_path)
        from langchain.text_splitter import RecursiveCharacterTextSplitter
        text_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=100)
        split_documents = text_splitter.create_documents([html_text])
    contents = [p.page_content for p in split_documents]
    
    sentence_embeddings = embed_sentences(contents)
    import numpy as np
    sentence_embeddings =np.reshape(sentence_embeddings,[len(contents),768])
    import faiss
    n_dimensions = sentence_embeddings.shape[1] #Number of dimensions (764)

    fastIndex = faiss.IndexFlatIP(n_dimensions) # We will create an index of type FlatL2, there are many kinds of indexes, you can look at it in their repo.
    sentence_embeddings =sentence_embeddings.astype('float32')
    faiss.normalize_L2(sentence_embeddings)
    fastIndex.add(sentence_embeddings) # Add the embedding vector to faiss index, it should of dtype 'float32'
    import time
    embedding_dim=768
    return fastIndex,contents
   
def get_context(query,index,contents):
  cost,res = do_faiss_lookup(index,query,m,20)
  relevant_documents = []
  for idx in res:
    relevant_documents.append(contents[idx].replace('\r\n','').replace('\n',''))
  context = ''.join(relevant_documents[:5])
  print(context)
  print(query)
  return get_new_prompt(context,query)
#   return "请根据文章回答问题，注意回答内容必须依据给定的文章,如果无法回答时候请回复根据您给定的文章无法回答问题，生成答案前请先思考验证一下是否满足需求。对于数字请严格根据文章来回答，不要自己编造，如果涉及计算请给出计算公式及结果由来。相关文章："+context+' 问题：'+query+' 回答:'
````

````python
from typing import Dict, Tuple
import os
import json
def parse_latest_plugin_call(text: str) -> Tuple[str, str]:
  i = text.rfind('<br>行动：')
  j = text.rfind('<br>行动输入:')
  k = text.rfind('<br>观察:')
  if 0 <= i < j:  # If the text has `Action` and `Action input`,
      if k < j:  # but does not contain `Observation`,
          # then it is likely that `Observation` is ommited by the LLM,
          # because the output text may have discarded the stop word.
          text = text.rstrip() + '<br>观察:'  # Add it back.
          k = text.rfind('<br>观察:')
  if 0 <= i < j < k:
      plugin_name = text[i + len('<br>行动：'):j].strip()
      plugin_args = text[j + len('<br>行动输入：'):k].strip()
      return plugin_name, plugin_args
  return '', ''
def parse_plugin_all(input_str):
    name,res = parse_latest_plugin_call(input_str)
    if name=='[Math]':
        try:
            res = res.strip().replace("'",'"')
            api_output = caculate_res = wolfram.run(json.loads(res)['query']).split('Answer:')[1].split('\n')[0]
            caculate_res = '观察:'+json.loads(res)['query']+'='+str(round(float(api_output.strip().strip('...')),4))
            import re 
            input_str = re.sub('观察:.*$',caculate_res,input_str)
            print(input_str)
            return input_str
        except Exception as e:
            print(e)
            pass 
    return ''
# input_str ="""思考：根据背景知识，需要计算2023年半年度的营业收入与2022年半年度的营业收入的差值。<br>行动：[Math]<br>行动输入: {"query": '1470499130.06-1389497988.98'}<br>观察: 171100134.08<br>思考：公司的营业收入同比增加了171100134.08。<br>最终答案：公司的营业收入同比增加了171100134.08。"""
# parse_plugin_all(input_str)
````

````python
def postprocess(self, y):
  if y is None:
      return []
  for i, (message, response) in enumerate(y):
      y[i] = (
          None if message is None else mdtex2html.convert((message)),
          None if response is None else mdtex2html.convert(response),
      )
  return y


gr.Chatbot.postprocess = postprocess


def parse_text(text):
    lines = text.split("\n")
    lines = [line for line in lines if line != ""]
    count = 0
    for i, line in enumerate(lines):
        if "```" in line:
            count += 1
            items = line.split("`")
            if count % 2 == 1:
                lines[i] = f'<pre><code class="language-{items[-1]}">'
            else:
                lines[i] = f"<br></code></pre>"
        else:
            if i > 0:
                if count % 2 == 1:
                    line = line.replace("`", "\`")
                    line = line.replace("<", "&lt;")
                    line = line.replace(">", "&gt;")
                    line = line.replace(" ", "&nbsp;")
                    line = line.replace("*", "&ast;")
                    line = line.replace("_", "&lowbar;")
                    line = line.replace("-", "&#45;")
                    line = line.replace(".", "&#46;")
                    line = line.replace("!", "&#33;")
                    line = line.replace("(", "&#40;")
                    line = line.replace(")", "&#41;")
                    line = line.replace("$", "&#36;")
                lines[i] = "<br>" + line
    text = "".join(lines)
    return text


task_history = []


def predict(query, chatbot,file):
    print('filename :'+str(file.name))
    print("User: " + parse_text(query))
    global faiss_index
    global contents 
    global input_filename
    if not input_filename or file.name != input_filename:
        faiss_index,contents = read_pdf_and_search(file.name)
        input_filename = file.name
    chatbot.append((parse_text(query), ""))
    fullResponse = ""
    first_prompt = get_context(query,faiss_index,contents)
    from transformers.trainer_utils import set_seed
    set_seed(1234)
    for response in model.chat_stream(tokenizer, first_prompt, history=task_history):
        chatbot[-1] = (parse_text(query), parse_text(response))

        yield chatbot
        fullResponse = parse_text(response)

    task_history.append((query, fullResponse))
    parse_full_text = parse_text(fullResponse)
    print("Qwen-7B-Chat: " + parse_text(fullResponse))
    if '<br>行动：' in parse_full_text:
        api_output = parse_plugin_all(parse_full_text)
        if api_output.strip()!='' :
            second_prompt = first_prompt+'\n'+api_output.replace('<br>','\n')
            print(first_prompt+'\n'+api_output.replace('<br>','\n'))
            fullResponse = ""
            from transformers.trainer_utils import set_seed
            set_seed(1234)
            for response in model.chat_stream(tokenizer, second_prompt, history=[]):
                chatbot[-1] = (parse_text(query), parse_text(response))
                yield chatbot
                fullResponse = parse_text(response)
            print("Qwen-7B-Chat: " + parse_text(fullResponse))

        


def regenerate(chatbot):
    if not task_history:
        yield chatbot
        return
    item = task_history.pop(-1)
    chatbot.pop(-1)
    yield from predict(item[0], chatbot)


def reset_user_input():
    return gr.update(value="")


def reset_state():
    task_history.clear()
    return []


with gr.Blocks() as demo:
    gr.Markdown("""<p align="center"><img src="https://modelscope.cn/api/v1/models/qwen/Qwen-7B-Chat/repo?Revision=master&FilePath=assets/logo.jpeg&View=true" style="height: 80px"/><p>""")
    gr.Markdown("""<center><font size=8>Qwen-7B-Chat Bot</center>""")
    gr.Markdown(
        """<center><font size=3>This WebUI is based on Qwen-7B-Chat, developed by Alibaba Cloud. (本WebUI基于Qwen-7B-Chat打造，实现聊天机器人功能。)</center>"""
    )
    gr.Markdown(
        """<center><font size=4>Qwen-7B <a href="https://modelscope.cn/models/qwen/Qwen-7B/summary">🤖 <a> | <a href="https://huggingface.co/Qwen/Qwen-7B">🤗</a>&nbsp ｜ Qwen-7B-Chat <a href="https://modelscope.cn/models/qwen/Qwen-7B-Chat/summary">🤖 <a>| <a href="https://huggingface.co/Qwen/Qwen-7B-Chat">🤗</a>&nbsp ｜ &nbsp<a href="https://github.com/QwenLM/Qwen-7B">Github</a></center>"""
    )

    chatbot = gr.Chatbot(lines=10, label='Qwen-7B-Chat', elem_classes="control-height")
    query = gr.Textbox(lines=2, label='Input')
    file =  gr.File(
                label='Upload your PDF/ Research Paper / Book here', file_types=['.pdf','.txt','.html']
            )

    with gr.Row():
        emptyBtn = gr.Button("🧹 Clear History (清除历史)")
        submitBtn = gr.Button("🚀 Submit (发送)")
        regenBtn = gr.Button("🤔️ Regenerate (重试)")

    submitBtn.click(predict, [query, chatbot,file], [chatbot], show_progress=True)
    submitBtn.click(reset_user_input, [], [query])
    emptyBtn.click(reset_state, outputs=[chatbot], show_progress=True)
    regenBtn.click(regenerate, [chatbot], [chatbot], show_progress=True)

    gr.Markdown(
        """<font size=2>Note: This demo is governed by the original license of Qwen-7B. We strongly advise users not to knowingly generate or allow others to knowingly generate harmful content, including hate speech, violence, pornography, deception, etc. (注：本演示受Qwen-7B的许可协议限制。我们强烈建议，用户不应传播及不应允许他人传播以下内容，包括但不限于仇恨言论、暴力、色情、欺诈相关的有害信息。)"""
    )

demo.queue().launch(share=True, inbrowser=True)
````
